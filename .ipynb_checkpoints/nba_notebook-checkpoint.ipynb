```python
class container : 
    def __init__(self,k,n,V,Q_max):
        self.k = k
        self.n = n
        self.mu = 0
        self.w = 0
        self.Q = 0
        self.Q_ = 0
        self.V = V
        self.Q_max = Q_max
        
    def backlog(self):
        if (self.Q_ <= self.V) : 
            self.Q = self.Q_
        else : 
            self.Q = self.Q_max
        
    def weight(self):
        self.w = math.log(1 + (self.Q/self.V))
        
    
    
    
```
